```python
# Read Antiplatelet1
Antiplatelet1 = pd.read_excel("C:/Users/brody/Desktop/School/Work/Internships/INOVA/Antiplatelet/Antiplatelet.xlsx", sheet_name=0, header=None)
Antiplatelet1 = pd.DataFrame(Antiplatelet1)
Antiplatelet1.columns = Antiplatelet1.iloc[0]
Antiplatelet1 = Antiplatelet1.iloc[1:]
Antiplatelet1.reset_index(drop=True, inplace=True)

# Drop rows with missing "Patient Number"
Antiplatelet1 = Antiplatelet1.dropna(subset=["Patient Number"])

# Reset the index
Antiplatelet1.reset_index(drop=True, inplace=True)

# Summary
Antiplatelet1_summary = Antiplatelet1.describe(include = 'all')
Antiplatelet1_summary
```

```python
# Read Antiplatelet2
Antiplatelet2 = pd.read_excel("C:/Users/brody/Desktop/School/Work/Internships/INOVA/Antiplatelet/Antiplatelet.xlsx", sheet_name=1, header=None,engine="openpyxl")

Antiplatelet2 = pd.DataFrame(Antiplatelet2)
Antiplatelet2.columns = Antiplatelet2.iloc[0]
Antiplatelet2 = Antiplatelet2.iloc[1:]
Antiplatelet2.reset_index(drop=True, inplace=True)

# Drop rows with missing "Patient Number"
Antiplatelet2 = Antiplatelet2.dropna(subset=["Patient Number"])

# Reset the index
Antiplatelet2.reset_index(drop=True, inplace=True)

#print(Antiplatelet2)

# Summary
Antiplatelet2_summary = Antiplatelet2.describe(include = 'all')
Antiplatelet2_summary
```

```python
# Read Antiplatelet3

Antiplatelet3 = pd.read_excel("C:/Users/brody/Desktop/School/Work/Internships/INOVA/Antiplatelet/Antiplatelet.xlsx", sheet_name=2, header=None,engine="openpyxl")

Antiplatelet3 = pd.DataFrame(Antiplatelet3)
Antiplatelet3.columns = Antiplatelet3.iloc[0]
Antiplatelet3 = Antiplatelet3.iloc[1:]
Antiplatelet3.reset_index(drop=True, inplace=True)

# Drop rows with missing "Patient Number"
Antiplatelet3 = Antiplatelet3.dropna(subset=["Patient Number"])

# Reset the index
Antiplatelet3.reset_index(drop=True, inplace=True)

# Summary
Antiplatelet3_summary = Antiplatelet3.describe()
Antiplatelet3_summary
```

```python
# Read Antiplatelet4

Antiplatelet4 = pd.read_excel("C:/Users/brody/Desktop/School/Work/Internships/INOVA/Antiplatelet/Antiplatelet.xlsx", sheet_name=3, header=None,engine="openpyxl")

Antiplatelet4 = pd.DataFrame(Antiplatelet4)
Antiplatelet4.columns = Antiplatelet4.iloc[0]
Antiplatelet4 = Antiplatelet4.iloc[1:]
Antiplatelet4.reset_index(drop=True, inplace=True)

# Drop rows with missing "Patient Number"
Antiplatelet4 = Antiplatelet4.dropna(subset=["Patient Number"])

# Reset the index
Antiplatelet4.reset_index(drop=True, inplace=True)

# Summary
Antiplatelet4_summary = Antiplatelet4.describe()
Antiplatelet4_summary
```

Merge Data

```python
# Merge datasets
Antiplatelet = pd.merge(Antiplatelet1, Antiplatelet2, on="Patient Number")
Antiplatelet = pd.merge(Antiplatelet, Antiplatelet3, on="Patient Number")
Antiplatelet = pd.merge(Antiplatelet, Antiplatelet4, on="Patient Number")


# Summary of merged dataset
print(Antiplatelet.describe(include = 'all'))

# Column names of merged dataset
print(Antiplatelet.columns)
```

Final Cleaning

```python
# Define NA strings
na_strings = ["NA", "N A", "N / A", "N/A", "N/ A", "Not Available", "NOt available", "Not documented", "Unknown", "Unavailable"]

# Replace NA strings with NaN
Antiplatelet.replace(na_strings, np.nan, inplace=True)

# Convert column 58 to string type
Antiplatelet.iloc[:, [58]] = Antiplatelet.iloc[:, [58]].astype(str)

replacement_mapping = {'0.0': '0', '1.0': '1', '2.0': '2', '3.0': '2', '4.0': '2', '5.0': '2'}

Antiplatelet['Initial.Antiplatelet.Strategy'] = Antiplatelet.iloc[:, [58]].replace(replacement_mapping)

#Antiplatelet.head(20)
```
